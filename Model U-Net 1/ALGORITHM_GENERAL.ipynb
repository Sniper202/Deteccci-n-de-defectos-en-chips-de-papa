```python
from keras.preprocessing.image import ImageDataGenerator
import time
import numpy as np

def data_augmentation(carp_raiz, carp_entrada, carp_salida, n_img_gen):
    dir_carp_entrada = os.path.join(carp_raiz, carp_entrada)
    num_cap_ = len(os.listdir(dir_carp_entrada))

    dir_carp_salida = os.path.join(carp_raiz, carp_salida)

    if os.path.exists(dir_carp_salida):
        print(f'El directorio {dir_carp_salida} ya existe!')
    if not os.path.exists(dir_carp_salida):
        os.makedirs(dir_carp_salida)
        print(f'El directorio {dir_carp_salida} se ha creado satisfactoriamente!')

    # Configuración de ImageDataGenerator para las transformaciones
    datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.3,
        height_shift_range=0.3,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest',
    )

    t_ini = time.time()
    for filename in os.listdir(dir_carp_entrada):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            dir_img_entrada = os.path.join(dir_carp_entrada, filename)
            img = image.load_img(dir_img_entrada, target_size=(1280, 1280))
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)

            img_aug_iter = datagen.flow(img_array, batch_size=1)

            for j in range(n_img_gen):
                img_augmented = next(img_aug_iter)[0].astype(np.uint8)
                img_aug_pil = image.array_to_img(img_augmented)

                # Ajustar la escala de los píxeles (0-255)
                img_aug_pil = img_aug_pil.convert('RGB')
                img_aug_pil.save(os.path.join(dir_carp_salida, f"{filename.split('.')[0]}_ag_{j}.{filename.split('.')[1]}"))

    t_fin = time.time()
    t_total = "{:.3f}".format(t_fin - t_ini)

    print('Proceso de aumento de datos completado!.')
    print(f'Total de img generadas: {num_cap_*n_img_gen}.')
    print(f'Tiempo: {t_total} s.')
```

```python
Carp_Raiz = 'D:/DETECCION_DE_DEFECTOS/ALGORITHMS_TESIS/ALG_1/DATA_A1'
Carp_img_ent = 'Data_random'
# Carp_Raiz = 'DATA_A_1'
# Carp_img_ent = 'Data_random'
# Carp_img_sal = 'Data_Augmentation'
Carp_img_sal = 'Data_Augmentation'
N_img_gen_x_img = 8
Variables =(Carp_Raiz,Carp_img_ent,Carp_img_sal,N_img_gen_x_img)

data_augmentation(Variables[0],Variables[1],Variables[2],Variables[3])
```

# Generar máscaras

```python
import cv2
# import keras.utils as image
def data_mask_generator(path_aux,output_dir_mask):
    
    for i in range(1):

        sample_class_aux = os.path.join(path_aux)# Crea la ruta a la carpeta de la clase trainnb
        img_path= os.listdir(sample_class_aux)
        for j in range(len(img_path)):

            img_list = os.path.join(sample_class_aux, img_path[j])

            img = image.load_img(img_list)
            img = image.img_to_array(img)
            img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            img_gray/=255.
            _, thr1= cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY)
            # Obtener el nombre de la imagen sin la extensión
            image_name = os.path.splitext(os.path.basename(img_list))[0]
            output_subdir = os.path.join(os.getcwd(), output_dir_mask)
            cv2.imwrite(os.path.join(output_subdir,f"{image_name}_mask.png"),thr1)
```

```python
output_dir_mask= "D:/DETECCION_DE_DEFECTOS/ALGORITHMS_TESIS/ALG_1/DATA_A1/Data_mask"
path_aux= "D:/DETECCION_DE_DEFECTOS/ALGORITHMS_TESIS/ALG_1/DATA_A1/Data_Nground"
# path_aux = "D:/DETECCION_DE_DEFECTOS/ALGORITHMS/Archivos_A1/Data_A_1/Data_Augmentation"
data_mask_generator(path_aux,output_dir_mask)
```

# Visualizar grupo de datos

```python
class PotatoChips_1:
    
    def __init__(self, name_title, name_xlabel, startpath, GeneralPath):
        self.name_title = name_title
        self.name_xlabel = name_xlabel
        self.startpath = startpath
        self.GeneralPath = GeneralPath
#         self.name_title2 = name_title2
        
        # pass
    
    def print_bold(self, text):
        print('\033[1m{}\033[0m'.format(text))
    
    def list_files(self):
        for root, dirs, files in os.walk(self.startpath):
            level = root.replace(self.startpath, '').count(os.sep)
            indent = ' '* 4 * (level)

            dir_name = '{}{}/'.format(indent, os.path.basename(root))
            if dir_name.strip().startswith('.'):
                continue

            self.print_bold('\n'+dir_name)

            subindent = ' ' * 4 * (level + 1)
            if level == 0:
                for j in files:
                    if j.startswith('.'):
                        continue
                        print('{}{}'.format(subindent, j))
            else:
                for i, j in enumerate(files):
                    print('{}{}'.format(subindent, j))
                    if i==2:
                        print('{}{}'.format(subindent, '...'))
                        break
                        
    def get_sample_counts(self):
        classes = []
        sample_counts = []

        for j in os.listdir(self.GeneralPath):
            train_class_path = os.path.join(self.GeneralPath, j)
            if os.path.isdir(train_class_path):
                classes.append(j)
                sample_counts.append(len(os.listdir(train_class_path)))
        return classes, sample_counts

    def plot_sample_counts(self, classes, sample_counts):
        plt.rcdefaults()
        fig, ax = plt.subplots(figsize=(7, 5))
        self.name_title
        y_pos = np.arange(len(classes))

        ax.barh(y_pos, sample_counts, align='center', color="green")
        ax.set_yticks(y_pos)
        ax.set_yticklabels(classes)
        ax.invert_yaxis()
        
        label_offset = max(sample_counts) * 0.001
        
        for i, count in enumerate(sample_counts):
            ax.text(count + label_offset, i, str(count), va='center', fontsize=8)
        
        ax.set_title(self.name_title)
        ax.set_xlabel(self.name_xlabel)
        plt.yticks(fontsize=10)
        plt.show()
    
    
    def plot_random_samples_class(self,classes,n_samples,name_title2):
        fig = plt.figure(figsize=(10,5))
        fig.suptitle(name_title2, fontsize = 13, y= .95, horizontalalignment = 'center',weight = 'bold')

        columns = n_samples
        rows = len(classes)
        for i, class_name in enumerate(classes):
            sample_class = os.path.join(self.GeneralPath, class_name)
            for j in range(1, n_samples + 1):
                fig.add_subplot(rows, columns, i * n_samples + j)
                plt.axis('off')

                # En la primera celda, muestra el nombre de la clase
                if j == 1:
                    plt.text(0.0, 0.5, class_name.replace(' ', '\n'), fontsize=11, wrap=True)
                    continue

                # Selecciona una imagen aleatoria de la clase actual
                random_img = os.path.join(sample_class, random.choice(os.listdir(sample_class)))

                # Carga la imagen, cambia su tamaño a 256x256 píxeles, normaliza los valores de píxeles
                img = image.load_img(random_img, target_size=(256, 256))
                img = image.img_to_array(img)
                img /= 255.

                # Muestra la imagen en la celda actual
                plt.imshow(img)

        # Presenta la figura completa con las muestras aleatorias
        plt.show() 
    
```
